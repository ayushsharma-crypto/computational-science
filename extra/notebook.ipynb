## Question 8:

The set of the given ODEs representing the Prey-Predator Model are as follows:

$$
\frac{dN_1(t)}{dt}=rN_1(t)\left [1-\frac{N_1(t)}{k} \right]-\alpha N_1N_2
$$
$$
\frac{dN_2(t)}{dt}=-cN_2(t) + \beta N_1N_2
$$

Where,

$N_1   \to$ Prey Population

$N_2   \to$ Predator Population

$\alpha \to$ Rate at which the predator captures its prey

$\beta \to$ Growth rate of the predator

$c \to$ Decay rate of the predator if $N_1 = 0$

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def simulate(initial_N1, initial_N2, params, time, intervals):
    N1 = initial_N1
    N2 = initial_N2
    r = params["r"]
    alpha = params["alpha"]
    beta = params["beta"]
    c = params["c"]
    k = params["k"]

    N1_list = []
    N2_list = []
    t_list = np.linspace(0, time, intervals)

    for t in t_list:
        N1_list.append(N1)
        N2_list.append(N2)
        dt = t_list[1] - t_list[0]

        ## dN = dN/dt * dt 

        dN1 =  (r*N1*(1-N1/k) - alpha*N1*N2) * dt
        dN2 = (beta*N1*N2 - c*N2) * dt

        N1 += dN1
        N2 += dN2


    return N1_list, N2_list, t_list
```

```
N1_init_list = [15, 12, 10]
N2_init_list = [3, 2, 2]
r_list = [1.5, 2.1, 2.5]
alpha_list = [0.4, 0.4, 0.5]
beta_list = [0.1, 0.15, 0.1]
c_list = [0.4, 0.5, 0.45]
k_list = [20, 25, 25]

l = len(N1_init_list)


for i in range(l):

    N1_init = N1_init_list[i]
    N2_init = N2_init_list[i]

    params = {}
    params["r"] = r_list[i]
    params["alpha"] = alpha_list[i]
    params["beta"] = beta_list[i]
    params["c"] = c_list[i]
    params["k"] = k_list[i]

    time = 100
    intervals = 500

    N1_list, N2_list, t_list = simulate(N1_init, N2_init, params, time, intervals)

    plt.figure(figsize=(15,10))
    plot_text = f"N1(0): {N1_init}\n"
    plot_text += f"N2(0): {N2_init}\n"
    plot_text += f"k: %.2f\n"%(params["k"])
    plot_text += f"r: %.2f\n"%(params["r"])
    plot_text += f"α: %.2f\n"%(params["alpha"])
    plot_text += f"β: %.2f\n"%(params["beta"])
    plot_text += f"c: %.2f"%(params["c"])

    plt.plot(t_list, N1_list, color='b')
    plt.plot(t_list, N2_list, color='r')

    plt.title("Prey-Predator Model")
    plt.ylabel("Population")
    plt.xlabel("Time Period")
    plt.legend(["Prey", "Predator"])
    plt.text(-25, 5,plot_text, fontsize=14)
    plt.subplots_adjust(left=0.25)
    plt.show()
```

The trajectories of the populations seem to fit the logic that
$$
\text{More Prey}\implies \text{More Predator}\implies \text{Less Prey}\implies \text{Less Predator}\implies \text{More Prey}
$$
This logic yeilds 2 oscialltory curves representing the 2 populations, where the curve for the prey polulation leads to curve for the predator population, which is evident from the plot.

The trajectories of the populations also show a convergent nature, i.e., the amplitudes of the oscillations seem to reduce as time passes. This is a direct consequence of the logistic nature of the growth of the prey population.
